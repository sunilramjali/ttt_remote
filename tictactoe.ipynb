2. The Game
(Clean version)

```python
# Run this code block to play the game

def box(): #create 3x3
    '''Create empty 3x3'''
    board = np.full((3,3), " ") #creates an empty 3x3 using numpy
    return board


def ref_board(): #visualised player guidance
    '''Visualised player guidance for board positions'''
    t = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] # Using my early visualisation of the matrix
    for row in t:
        for value in row:
            print(value, end = " | ")
        print()
        print("Board positions:")
        print("-" * 11)


def viz(board, turn): #visualised box that updates upon player action
    '''Updating visualised box'''
    print(f"\nTurn {turn}\n")

    for i, row in enumerate(board): # rows used for enumerate
        print(" " + " | ".join(row)) # | and - for line visualisation in between boxes
        if i < 2:
            print("-" * 9)


def win_con(board): #determine winning condition
    '''Winning condition'''
    # rows
    for i, row in enumerate(board): #i is represented as row, and using enumerate to refer and count
        if row[0] != " " and row[0] == row[1] == row[2]: # also, takes in account for empty cells, so empty cells cannot be determined as a winner!
            print(row[0], "wins on row!", i)
            return True
        
    # cols
    for j in range(3): # using range for columns, and reffered as j
        if board[0][j] != " " and board[0][j] == board[2][j]:
            print(board[0][j], "wins on column", j)
            return True
    
    # diagonals
    if board[i][0] != " " and board[0][0] == board[1][1] == board [2][2]:
        print(board[0][0], "wins on diagonal!")
        return True

    elif board[i][0] != " " and board[0][2] == board[1][1] == board[2][0]:
        print(board[0][2], "wins on diagonal!")
        return True
    
    return False # breaks if none of the condition is made


def player_input(board, symbol):# player inputs
    '''Player Inputs'''
    while True:
        try:
            move = int(input(f"{symbol} move (1-9): ")) # instruction 1 to 9

            if move not in range(1,10): # checks the box limit
                print("invalid position. Use numbers 1 to 9.")
                continue

            row = (move -1) // 3 # converting rows and columns to 1 to 9
            col = (move -1) % 3

            if board[row][col] != " ": # checks if the spot is taken or not
                print('That spot is already taken!')
                continue

            board[row][col] = symbol 
            break

        except ValueError:
            print("invalid input. Enter a number form 1 to 9.")


def draw_outcome(board): # creating outcome for a draw
    return " " not in board


def game(): # loops and prinitng results
    '''Loops and Print Results'''
    board = box()
    current_player = "X"
    turn = 1

    ref_board()

    while True: # loop
        viz(board, turn)
        player_input(board, current_player)

        if win_con(board): # prints after winning condition
            viz(board, turn)
            print(f"Game over! {current_player} wins.")
            break

        if draw_outcome(board): # condition for draw
            viz(board, turn)
            print("Game over! It's a draw.")
            break

        current_player = "O" if current_player == "X" else "X" # alternates player's turn
        turn += 1 # increments after each turn


game()
```
