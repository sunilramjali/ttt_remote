```python
# Visualising row and column indicator for tic tac toe

def viz():
    t = np.empty((3,3), dtype = object)

    for i in range(3):
        for j in range(3):
            t[i,j] = (i,j)
    print(t)

viz()
```

```python
# Creating winning conditions

def win_con (board):
    '''Winning Conditions Check'''

    # check rows
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2]:
            print(board[i][0], "wins on row", i)
            return
        
    # check columns
    for j in range(3):
        if board[0][j] == board[1][j] == board[2][j]:
            print(board[0][j], "wins on column", j)
            return
    
    #check diagonals
    if board[0][0] == board[1][1] == board [2][2]:
        print(board[0][0], "wins on diagonal")

    elif board[0][2] == board[1][1] == board[2][0]:
        print(board[0][2], "wins on diagonal")

    else:
        print("The match is draw")
    
board = box()
win_con(board)
```
