2. The Game
(Clean version)

```python
# Run this code block to play the game

if "stats" not in globals(): # prevents resetting stats after every run
    stats = { # using dictionary to efficiently store all the statitsics and count using enumerate
        "Games played": 0,
        "X wins": 0,
        "O wins": 0,
        "Draws": 0,
        "Quickest Win": None
}


#Global player stats
if "player_stats" not in globals():
    player_stats = {}

def box(): #create 3x3
    '''Create empty 3x3'''
    board = np.full((3,3), " ") #creates an empty 3x3 using numpy
    return board


def ref_board(): #visualised player guidance
    '''Visualised player guidance for board positions'''
    t = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] # Using my early visualisation of the matrix
    print("Board positions:")
    for row in t:
        for value in row:
            print(value, end = " | ")
        print()
        print("-" * 11)


def viz(board, turn = None): #visualised box that updates upon player action
    '''Updating visualised box'''
    if turn is not None: # for accurate turn counting, without this there would be an empty turn!
        print(f"\nTurn {turn}\n")

    for i, row in enumerate(board): # rows used for enumerate
        print(" " + " | ".join(row)) # | and - for line visualisation in between boxes
        if i < 2:
            print("-" * 9)


def win_con(board): #determine winning condition
    '''Winning condition'''
    # rows
    for i, row in enumerate(board): #i is represented as row, and using enumerate to refer and count
        if row[0] != " " and row[0] == row[1] == row[2]: # also, takes in account for empty cells, so empty cells cannot be determined as a winner!
            return True
        
    # cols
    for j in range(3): # using range for columns, and reffered as j
        if board[0][j] != " " and board[0][j] == board[1][j] == board[2][j]:
            return True
    
    # diagonals
    if board[0][0] != " " and board[0][0] == board[1][1] == board [2][2]:
        return True

    elif board[0][2] != " " and board[0][2] == board[1][1] == board[2][0]:
        return True
    
    return False # breaks if none of the condition is made


def player_input(board, symbol, player_name):# player inputs
    '''Player Inputs'''
    while True:
        try:
            move = int(input(f"{player_name}{symbol} move (1-9): ")) # instruction 1 to 9

            if move not in range(1,10): # checks the box limit
                print("invalid position. Use numbers 1 to 9.")
                continue

            row = (move -1) // 3 # converting rows and columns to 1 to 9
            col = (move -1) % 3

            if board[row][col] != " ": # checks if the spot is taken or not
                print('That spot is already taken!')
                continue

            board[row][col] = symbol 
            break

        except ValueError: # for invalid inputs
            print("invalid input. Enter a number form 1 to 9.")


def draw_outcome(board): # check if board is full
    '''Check if board is full'''
    return " " not in board


def game_stats():# recording and calculation stats
    '''Statistics'''
    print("\nGame Statistics")
    print("-" * 15)

    for i, (key, value) in enumerate(stats.items(), start = 1):
        print(f"{i}. {key}: {value}") 


def players_reg():# player registry
    '''Register player names and initialise stats'''
    player_x = input("Enter player name for X: ").strip()
    player_o = input("Enter player name for O:").strip()

    while player_x == "":
        player_x = input("Player X name cannot be empty. enter player name for X: ").stip() # input player names while removing spaces infornt or end of their name

    while player_o == "":
        player_o = input("Player O name cannot be empty. enter player name for O: ").stip()

    for player in [player_x, player_o]:
        if player not in player_stats:
            player_stats[player] = {
                "Games": 0,
                "Wins": 0,
                "Losses": 0,
                "Draws": 0,
            }

    return player_x, player_o

def player_statistics():# print player stats
    '''Display all player statistics'''
    print("\nPlayer Statistics")
    print("-" * 20)

    for player, data in player_stats.items():
        print(f"\n{player}")
        for key, value in data.items():
            print(f"{key}: {value}")


def game(): # loops and printing results
    '''Loops and Print Results'''
    board = box()
    current_player = "X"
    turn = 1

    player_x, player_o = players_reg() #store data

    stats["Games played"] += 1 #stats increment
    player_stats[player_x]["Games"] += 1
    player_stats[player_o]["Games"] += 1

    ref_board()
    viz(board) # Added outside the loop, after testing so it does not count as a turn

    while True: # loop
        current_name = player_x if current_player == "X" else player_o # Turn indication

        player_input(board, current_player, current_name) # asks for input
        viz(board, turn) # prints visualised board, is under

        if win_con(board): # prints after winning condition, also added more and less than 5, for a quick fix for game ending at
            winner = player_x if current_player == "X" else player_o
            loser = player_o if current_player == "X" else player_x
            
            if current_player == "X": # tallying wins
                stats["X wins"] += 1
            else:
                stats["O wins"] += 1

            player_stats[winner]["Wins"] += 1
            player_stats[loser]["Losses"] += 1

            if stats["Quickest Win"] is None or turn < stats["Quickest Win"]: # update quickest win
                stats["Quickest Win"] = turn

            print(f"\nGame over! {winner} ({current_player}) wins!!!")
            break

        if draw_outcome(board): # condition for draw
            stats["Draws"] += 1
            player_stats[player_x]["Draws"] += 1
            player_stats[player_o]["Draws"] += 1

            print("\nGame over! It's a draw.")
            break

        current_player = "O" if current_player == "X" else "X" # alternates player's turn
        turn += 1 # increments after each turn

    game_stats()
    player_statistics()


while True:
    game()

    again = input("\nPlay another game? (y/n): ").strip().lower() # option to start again
    if again != "y":
        print("\nFinal Statistics:")
        game_stats()
        player_statistics()
        print("Thanks for playing!")
        break
```
