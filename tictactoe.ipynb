```python
# Import numpy package to create a matrix for tic tac toe box

import numpy as np
```

```python
# Visualising row and column indicator for tic tac toe

def viz():
    t = np.empty((3,3), dtype = object)

    for i in range(3):
        for j in range(3):
            t[i,j] = (i,j)
    print(t)

viz()
```
