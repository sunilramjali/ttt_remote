2. The Game
(Clean version)

```python
# Run this code block to play the game
def box_beta2():
    board = np.full((3,3), " ")
    return board

def print_board(board):
    for row in board:
        print(" " + " | ".join(row))
        print("-" * 9)

def win_con(board):
    for i in range(3):
        if board[i][0] != " " and board[i][0] == board[i][1] == board[i][2]:
            print(board[i][0], "wins on row", i)
            return True
        
    for j in range(3):
        if board[0][j] != " " and board[0][j] == board[1][j] == board[2][j]:
            print(board[0][j], "wins on column", j)
            return True
    
    if board[0][0] != " " and board[0][0] == board[1][1] == board[2][2]:
        print(board[0][0], "wins on diagonal")
        return True

    if board[0][2] != " " and board[0][2] == board[1][1] == board[2][0]:
        print(board[0][2], "wins on diagonal")
        return True

    return False

def player_input(board, symbol):
    while True:
        try:
            row, col = map(int, input(f"{symbol} move (row col): ").split())

            if row not in range(3) or col not in range(3):
                print("Invalid position. Use numbers 0 to 2.")
                continue

            if board[row][col] != " ":
                print("That spot is already taken.")
                continue

            board[row][col] = symbol
            break

        except ValueError:
            print("Invalid input. Type two numbers like: 1 2")

board = box_beta2()
current = "X"

for turn in range(9):
    print_board(board)
    player_input(board, current)

    if win_con(board):
        print_board(board)
        break

    current = "O" if current == "X" else "X"
else:
    print_board(board)
    print("Draw")
```

```python
# Creating winning conditions

def win_con_alpha (board):
    '''Winning Conditions Check'''

    # check rows
    for i in range(3):
        if board[i][0] != " " and board[i][0] == board[i][1] == board[i][2]: # [i][j] being [rows][columns] in 3x3 matrix, and also added check for empty not being counted as winner!
            print(board[i][0], "wins on row", i)
            return
        
    # check columns
    for j in range(3):
        if board[i][0] != " " and board[0][j] == board[1][j] == board[2][j]:
            print(board[0][j], "wins on column", j)
            return
    
    #check diagonals
    if board[i][0] != " " and board[0][0] == board[1][1] == board [2][2]:
        print(board[0][0], "wins on diagonal")

    elif board[i][0] != " " and board[0][2] == board[1][1] == board[2][0]:
        print(board[0][2], "wins on diagonal")

    else:
        print("The match is draw")
    
board = box_rand() # random X and O
win_con_alpha(board)
```

```python
# creating an empty board

def box_beta():
    '''Emppty board (base)'''
    board = np.full((3, 3), " ") # np.empty caused an issue, as empty inputs random value!
    print(board)
    return board

box_beta()
```

```python
# printing visualised baord

def viz_beta(board):
    '''Board Visualisation'''
    for row in board:
        print(" " + " | ".join(row))
        print("-" * 9)

viz_beta(board)
```

```python
# creating player input

def player_input_alpha(board, symbol):
    '''Player Inputs'''

    while True:
        try:
            row, col = map(int, input(f"{symbol} move (row col): ").split()) # player instruction

            # check board limits
            if row not in range(3) or col not in range(3):  # creates barrier/limit 3x3 matrix
                print("Invalid position. Use numbers 0-2.")
                continue

            # check for empty cells
            if board[row][col] != " ":
                print("The spot is taken!!")
                continue

            board[row][col] = symbol
            break

        except:
            print("Invalid input. Example: 1 2")

# testing player_input_alpha

board = box_beta()
viz_beta(board)

player_input_alpha(board, "X")
viz_beta(board)

player_input(board, "O")
viz_beta(board)
```

The input seems to work fine. Constructing looping now.

```python
# looking at win_con_alpha

def win_con_beta (board):
    '''Winning Conditions Check'''

    # check rows
    for i in range(3):
        if board[i][0] != " " and board[i][0] == board[i][1] == board[i][2]: # [i][j] being [rows][columns] in 3x3 matrix, and also added check for empty not being counted as winner!
            print(board[i][0], "wins on row", i)
            return True
        
    # check columns
    for j in range(3):
        if board[i][0] != " " and board[0][j] == board[1][j] == board[2][j]:
            print(board[0][j], "wins on column", j)
            return True
    
    #check diagonals
    if board[i][0] != " " and board[0][0] == board[1][1] == board [2][2]:
        print(board[0][0], "wins on diagonal")
        return True

    elif board[i][0] != " " and board[0][2] == board[1][1] == board[2][0]:
        print(board[0][2], "wins on diagonal")
        return True
    
    return False

# creating stop after filling the board

current = "X" 

for turn in range(9): # only moves as 18 boxes, so 9 turns per player
    viz_beta(board)
    player_input_alpha(board, current)

    if win_con_alpha(board): # stops when a player reaches a win condition
        viz_beta(board)
        break

    current = "O" if current == "X" else "X" # switches player

else: # when no wining condition met, it is concluded as draw
    viz_beta(board) 
    print("The match is Draw")
```
